```python
#import dependencies and libraries
import numpy as np
import pandas as pd
```

```python
#first import the original csv and data file
file_path = "Resources\mental_health_dataset.csv"
df_mental_health = pd.read_csv(file_path)
#convert timestamp to datetime
df_mental_health['Timestamp'] = pd.to_datetime(df_mental_health['Timestamp'])
```

```python
#checking the initial data
df_mental_health.head()
```

```python
#then clean the data to drop all of the nan values
df_cleaned_mental_health = df_mental_health.dropna()
```

```python
#checking to make sure data is cleaned
df_cleaned_mental_health.head()
```

```python
#export the cleaned data back into the resources folder called cleaned_mental_dataset and have it have encoding of utf-8
df_cleaned_mental_health.to_csv("Resources\cleaned_mental_dataset.csv", encoding='utf-8', index=False)

```
